```python
def expand_boundaries(text: str, spans: List[List]):
    punctuations = set(string.punctuation)
    punctuations.add('’')
    punctuations.add('“')
    punctuations.add('”')
    punctuations.add("—")
    n = len(text)

    def expand_left(start):
        while start >= 0 and not text[start].isspace() and text[start] not in punctuations:
            start -= 1
        return start + 1

    def expand_right(end):        
        while end <= n and not text[end - 1].isspace() and text[end - 1] not in punctuations:
            end += 1
        return end - 1
    exception_list = {"Chinese", "Taiwanese", "Malaysian", "European", "Asian", "South African"}

    expanded_spans = []
    for span in spans:
        s, e, label = span
        s_expanded = expand_left(s - 1)
        e_expanded = expand_right(e + 1)

        new_span = [s_expanded, e_expanded, label]
        
        if text[s_expanded:e_expanded] in exception_list:
            print(f"Skipping {text[s:e]} label because it is part of {text[s_expanded:e_expanded]} token present in "
                  f"exception list") 
            continue

        if s != s_expanded or e != e_expanded:
            print(f"Changed {text[s:e]} to {text[s_expanded:e_expanded]}")

        expanded_spans.append(new_span)

    return expanded_spans


def strip_trailing_spaces(text: str, spans: List[List]):
    def trim_left(start):
        while text[start].isspace():
            start += 1
        return start

    def trim_right(end):
        while text[end - 1].isspace():
            end -= 1        
        return end

    trimmed_spans = []
    for span in spans:
        s, e, label = span
        s_trimmed = trim_left(s)
        e_trimmed = trim_right(e)

        new_span = [s_trimmed, e_trimmed, label]

        if s != s_trimmed or e != e_trimmed:
            print(f"Changed {text[s:e]} to {text[s_trimmed:e_trimmed]}")

        trimmed_spans.append(new_span)

    return trimmed_spans

def remove_unnecessary_predix_and_suffix(text, spans):              
    cleaned_spans = []
    for span in spans:
        s, e, label = span
        start = s
        end = e 
        curr_span = text[s:e]
        if curr_span[-2:] in {"\'s", "s'", "’s"}:
            end -= 2               
        if curr_span[:3] in {"Mr ", "Ms "}:
            start += 3              
        if curr_span[:4] in {"Mr. ", "Ms. ", "Mrs ", "The "}:
            start += 4                
        if curr_span[:5] in {"Mrs. ", "Miss "}:
            start += 5                    
        if s != start or e != end:
            print(f"Changed {text[s:e]} to {text[start:end]}")
        cleaned_spans.append([start, end, label])
    
    return cleaned_spans


def add_position_suffixes(text, spans):
    positioned_spans = []
    for span in spans:        
        s, e, label = span

        curr_span = text[s:e]
        tokenized_span = []
        start = s
        doc = nlp(curr_span)
        is_start = True        
        for sent in doc.sentences:
            for tok in sent.tokens:                
                label_new = []
                if is_start:
                    for lab in label:                        
                        label_new.append(f"B-{lab}")
                    tokenized_span.append([s + tok.start_char, s + tok.end_char, label_new])
                    is_start = False
                else:
                    for lab in label:
                        label_new.append(f"I-{lab}")
                    tokenized_span.append([s + tok.start_char, s + tok.end_char, label_new])                                        

        positioned_spans.extend(tokenized_span)

    return positioned_spans

def group_spans(spans):
    entities = {"PER", "LOC", "ORG", "PRODUCT"}

    entity_spans = []
    rel_spans = []
    for span in spans:
        if span[2] in entities:
            entity_spans.append(span)
        else:
            rel_spans.append(span)

    entity_grps = {}
    for span in entity_spans:
        s, e, label = span
        if (s, e) in entity_grps:
            entity_grps[s, e].append(label)
        else:
            entity_grps[s, e] = [label]

    rel_grps = {}
    for span in rel_spans:
        s, e, label = span
        if (s, e) in rel_grps:
            rel_grps[s, e].append(label)
        else:
            rel_grps[s, e] = [label]

    entity_grps = {k:v for k,v in sorted(entity_grps.items(), key=lambda x:x[0][0])}
    rel_grps = {k:v for k,v in sorted(rel_grps.items(), key=lambda x:x[0][0])}

    entity_grps = [[s, e, labels] for (s, e), labels in entity_grps.items()]
    rel_grps = [[s, e, labels] for (s, e), labels in rel_grps.items()]

    return entity_grps, rel_grps


def add_o_tokens(text, spans):

    def add_o_tokens_to_str(s, offset):
        o_spans = []
        doc = nlp(s)
        for sentence in doc.sentences:
            for token in sentence.tokens:            
                o_spans.append([token.start_char + offset, token.end_char + offset, ["O"]])
                assert token.text == text[token.start_char + offset:token.end_char + offset]
        return o_spans
    
    updated_spans = []
    i = 0
    for span in spans:
        s, e, label = span
        unlabelled_str = text[i:s]
        updated_spans.extend(add_o_tokens_to_str(unlabelled_str, i))        
        updated_spans.append(span)
        i = e
    
    updated_spans.extend(add_o_tokens_to_str(text[i:], i))
    recreate_text(text, updated_spans)
    return updated_spans

def clean_span_overlaps(text, spans):
    updated_spans = []
    sorted_spans = sorted(spans, key=lambda x:x[0])

    prev_s, prev_e = -1, -1
    for span in sorted_spans:
        if len(updated_spans) > 0:
            prev_s, prev_e, prev_label = updated_spans[-1]            
        s, e, label = span
        if s < prev_e:
            if e - s > prev_e - prev_s:
                updated_spans[-1] = [s, e, label]
                print(f"Discarded {text[prev_s:prev_e], prev_label} and used {text[s:e], label}")
            else:
                print(f"Discarded {text[s:e], label} and used {text[prev_s:prev_e], prev_label}")        
        else:
            updated_spans.append(span)
    
    return updated_spans



def recreate_text(text, spans):
    s1 = ""
    s2 = ""
    for c in text:
        if not c.isspace():
            s1 += c
    
    for span in spans:
        s, e, _ = span
        token = text[s:e]
        for c in token:
            assert not c.isspace(), f"Encountered space in {token, span}, which is not expected in labels after -B -I separation"
            s2 += c 

    assert s1 == s2, f"Strings are not matching, some characters have been added or removed. \ns1 = {s1}\n s2 = {s2}\n len(s1) = {len(s1)} len(s2) = {len(s2)}"
```

# Remove trailing spaces

```python
print("Removing trailing whitespaces")
gold_data.label = gold_data.apply(lambda x: strip_trailing_spaces(x.data, x.label), axis=1)
```

# Extend boundaries

```python
print("Extending boundaries")
gold_data.label = gold_data.apply(lambda x: expand_boundaries(x.data, x.label), axis=1)
```

# Clean prefix and suffixes

```python
print("Cleaning unnecessary prefix and suffixes")
gold_data.label = gold_data.apply(lambda x: remove_unnecessary_predix_and_suffix(x.data, x.label), axis=1)
```

# Groupping multiple labels 

```python
applied_df = gold_data.apply(lambda x:group_spans(x.label), axis=1, result_type='expand')
applied_df.columns=['entities', 'relationships']
gold_data = pd.concat((gold_data, applied_df), axis=1)
```

# Remove overlaps 

```python
gold_data.relationships = gold_data.apply(lambda x:clean_span_overlaps(x.data, x.relationships), axis=1)
gold_data.entities = gold_data.apply(lambda x:clean_span_overlaps(x.data, x.entities), axis=1)
```

# Add B and I information in tokens

```python
from tqdm import tqdm
tqdm.pandas()

gold_data.relationships = gold_data.progress_apply(lambda x:add_position_suffixes(x.data, x.relationships), axis=1)
gold_data.entities = gold_data.progress_apply(lambda x:add_position_suffixes(x.data, x.entities), axis=1)
```

# Create token level data

```python
def create_token_level_data(text, spans):
    rows = []
    for span in spans:
        s, e, label = span
        token = text[s:e]
        rows.append((token, label))
    return rows 

entity_df = pd.DataFrame([], columns={'doc_idx', 'sent_id', 'gold_token', 'gold_label'})
rel_df = pd.DataFrame([], columns={'doc_idx', 'sent_id', 'gold_token', 'gold_label'})
for i, row in gold_data.iterrows():
    sub_entity_df = pd.DataFrame(create_token_level_data(row.data, row.entities), columns=['gold_token', 'gold_label'])
    sub_rel_df = pd.DataFrame(create_token_level_data(row.data, row.relationships), columns=['gold_token', 'gold_label'])

    sub_entity_df['doc_idx'] = i
    sub_rel_df['doc_idx'] = i 

    entity_df = pd.concat((entity_df, sub_entity_df), axis=0)
    rel_df = pd.concat((rel_df, sub_rel_df), axis=0)

entity_df.reset_index(inplace=True, drop=True)
rel_df.reset_index(inplace=True, drop=True)
```

# Create sentence level raw data

```python
def create_sentences(text):
    doc = nlp(text)
    sents = []
    for sent in doc.sentences:
        sents.append(sent.text)
    return sents

raw_df = pd.DataFrame([], columns=['doc_idx', 'text'])
for i, row in tqdm(gold_data.iterrows(), total=len(gold_data)):
    sub_raw_df = pd.DataFrame(create_sentences(row.data), columns=['text'])
    sub_raw_df['doc_idx'] = i

    raw_df = pd.concat((raw_df, sub_raw_df), axis=0)

raw_df.reset_index(inplace=True)
raw_df.rename(columns={'index': 'sent_idx'}, inplace=True)
```

# Assess the mergability of raw text data and token level data 

```python
raw_grps = raw_df.groupby(by=['doc_idx'])
raw_grps = {k:v for k,v in raw_grps}

entity_grps = entity_df.groupby(by=['doc_idx'])
entity_grps = {k:v for k,v in entity_grps}

for k, v in raw_grps.items():
    s1 = ""
    for sen in v.text:
        for c in sen:
            if not c.isspace():
                s1 += c
    
    v1 = entity_grps[k]
    s2 = ""
    for word in v1.gold_token:
        for c in word:
            assert not c.isspace()
            s2 += c
    
    assert s1 == s2, f"Failed for doc_id {k}"
```
